```python
import cv2
import pytesseract
from PIL import Image
import numpy as np
from sklearn.cluster import KMeans
from collections import Counter
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from io import BytesIO
import base64

from PIL import Image 
from pytesseract import pytesseract 
import cv2
import pytesseract
from PIL import Image
import random
import os
import pandas as pd
```

```python
def extract_color(image_path, k=8):
    path_to_tesseract = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
    # Read the image using OpenCV
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Unable to load image from {image_path}")

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Reshape the image to be a list of pixels
    pixels = image.reshape((-1, 3))

    # Use KMeans clustering to find dominant colors
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pixels)
    dominant_colors = kmeans.cluster_centers_.astype(int)

    return dominant_colors
def generate_random_font():
    path_to_tesseract = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
    # Simulate generating random font information
    possible_fonts = ["Arial", "Times New Roman", "Courier New", "Verdana", "Comic Sans MS"]
    random_font = random.choice(possible_fonts)
    random_font_size = random.randint(10, 30)

    return {
        'font_type': random_font,
        'font_size': random_font_size
    }

def extract_text_and_font(image_path):
    # Read the image using OpenCV
    image = cv2.imread(image_path)
    path_to_tesseract = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
    if image is None:
        print(f"Error: Unable to load image from {image_path}")
        return None

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Use Tesseract OCR to extract text from the image
    text = pytesseract.image_to_string(Image.fromarray(gray))
     # Count the number of words
    word_count = len(text.split())
    # Generate random font information

    return text, word_count


def create_dataset(root_folder):
    data = []

    # Loop through each subfolder in the root folder
    for subfolder in os.listdir(root_folder):
        subfolder_path = os.path.join(root_folder, subfolder)

        if os.path.isdir(subfolder_path):
            # Loop through each image in the subfolder
            for filename in os.listdir(subfolder_path):
                if filename.endswith(('.jpg', '.png', '.jpeg', '.jfif')):
                    image_path = os.path.join(subfolder_path, filename)

                    # Extract features from the image
                    dominant_colors = extract_color(image_path)
                    random_font_info = generate_random_font()
                    text, word_count = extract_text_and_font(image_path)

                    # Append the features to the dataset
                    data.append({
                        'client': subfolder,
                        'Image_Path': image_path,
                        'Dominant_Colors': dominant_colors,
                        'Text_Content': text,
                        'Word_Count': word_count,
                        'font_type': random_font_info['font_type'],
                        'font_size': random_font_info['font_size']
                    })

    # Convert the list of dictionaries to a DataFrame
    dataset = pd.DataFrame(data)
    return dataset

# Example usage
image_folder = r"BatoolTestImg"

# Create the dataset
full_dataset = create_dataset(image_folder)

# Save the dataset to a CSV file
# full_dataset.to_csv("extracted_features_dataset.csv", index=False)
```

```python
def encode_image_to_base64(Image_Path):
    with open(Image_Path, 'rb') as file:
        base64_bytes = base64.b64encode(file.read())
        base64_string = base64_bytes.decode()
        return base64_string

df['base64'] = df['Image_Path'].apply(encode_image_to_base64)
```

```python
client = OpenAI()

def vision_api_describe_image(url):
    response = openai.ChatCompletion.create(
        model="gpt-4-vision-preview",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "image_url",
                        "image_url": {
              "url": f"data:image/jpeg;base64,{url}"
            }
                    },
                    {
                        "type": "text",
                        "text": "Describe the image in one world for each (lighting,theme of graphic, style) as table"
                    }
                ]
            }
        ],
        max_tokens=300
    )
    # accessing the content
    description_text = response.choices[0].message.content
    return description_text

```

```python
df['description'] = df['base64'].apply(vision_api_describe_image)
```

```python
def parse_description(description, standard_columns):
    lines = description.split('\n')[2:]
    description_dict = {col: None for col in standard_columns}  # Initialize with standard columns
    for line in lines:
        parts = line.split('|')
        if len(parts) >= 3:  # To ensure there are enough parts in the line
            aspect, descriptor = parts[1].strip(), parts[2].strip()
            if aspect in standard_columns:  # Only add if aspect is in the standard columns
                description_dict[aspect] = descriptor
    return description_dict

# Standard columns you want to extract
standard_columns = ['Lighting', 'Theme', 'Style']


# Apply the parsing function
parsed_descriptions = df['description'].apply(lambda x: parse_description(x, standard_columns))

# Convert to DataFrame and join
description_df = pd.DataFrame.from_records(parsed_descriptions)
result_df = df.join(description_df).drop(columns=['description'])

# Display the updated DataFrame
result_df
```
